### The Asterisk (*var)
When used with an asterisk prefix, the syntax has two primary functions: packing and unpacking.

#### 1. Packing (within function definitions)
In a function definition, *var (commonly seen as *args) is used to pack an arbitrary number of positional arguments into a single tuple within the function. This allows a function to accept a variable number of inputs. 

```python
def function_with_packing(*numbers):
    # 'numbers' becomes a tuple of all arguments passed
    total = sum(numbers)
    print(f"Sum: {total}, Type: {type(numbers)}")

function_with_packing(1, 2, 3, 4)
# Output: Sum: 10, Type: <class 'tuple'>
```

#### 2. Unpacking (within function calls or assignments)
In a function call or an assignment, the asterisk is used to unpack elements from an iterable (like a list or tuple).

```python
# In a function call or an assignment, the asterisk is used to unpack elements from an iterable (like a list or tuple).
my_list = [10, 20, 30]
# Unpacks my_list into three arguments: 10, 20, 30
function_with_packing(*my_list)
# Output: Sum: 60, Type: <class 'tuple'>
```

```python
# Assignment: It can be used to assign values from a sequence to multiple variables in a single line, 
# especially useful for catching multiple values.

head, *tail = [1, 2, 3, 4, 5]
# head is assigned 1, tail is assigned [2, 3, 4, 5]
print(f"Head: {head}, Tail: {tail}")
# Output: Head: 1, Tail: [2, 3, 4, 5]
```

### Python Tuples

Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called.

```python
pytuple = ("apple", "banana", "cherry")
print(pytuple)

# Tuples allow duplicate values:
thistuple = ("apple", "banana", "cherry", "apple", "grapes", "orange", "kiwi", "melon", "mango")
print(thistuple)

# Access Tuples
print(thistuple[1])

## Negative Indexing : Start from end
print(thistuple[-1])

## Range of Indexes : Note: The search will start at index 2 (included) and end at index 5 (not included).
print(thistuple[2:5])
## returns the items from the beginning to, but NOT included, "grapes":
print(thistuple[:4])

## returns the items from "cherry" and to the end:
print(thistuple[2:])

# Tuples Length
print(len(thistuple))

#
tuple1 = ("abc", 34, True, 40, "male")
print(type(tuple1))

#
# Change Tuple Values
#
tuple2 = list(tuple1)
tuple2[0] = "kiwi"
tuple1 = tuple(tuple2)

print(tuple1)

# Add Items
tuple2 = list(tuple1)
tuple2.append("orange")
tuple1 = tuple(tuple2)
print(tuple1)

# Add tuples  Or Merge Tuples
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y

print(thistuple)

#
# Unpack Tuples
#
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

# Using Asterisk *
(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)

# If the asterisk is added to another variable name than the last, Python will assign values to the variable until the number of values left matches the number of variables left.
(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

#### Walrus operator:

```python
sample_data = [
    {"userId": 1, "name": "rahul", "completed": False},
    {"userId": 1, "name": "rohit", "completed": False},
    {"userId": 1, "name": "ram", "completed": False},
    {"userId": 1, "name": "ravan", "completed": True}
]

#
#
#
print("With Python 3.8 Walrus Operator:")
for entry in sample_data:
    if name := entry.get("name"):
        print(f'Found name: "{name}"')

#
#
#
print("--"*10)
print("Without Walrus operator:")
for entry in sample_data:
    name = entry.get("name")
    if name:
        print(f'Found name: "{name}"')
```

#### *args & **kwargs

```python
#Explanation:
## *args: collects positional arguments into a tuple.
## **kwargs: collects keyword arguments into a dictionary.

def student_info(*args, **kwargs):
    print("Subjects:", args)        # Positional arguments
    print("Details:", kwargs)       # Keyword arguments

# Passing subjects as *args and details as **kwargs
student_info("Math", "Science", "English", Name="Alice", Age=20, City="New York")
```

#### Q-001. If this value is set to a number other than 200, the number of tasks in the shuffle stage will reflect that new value.

If this value is set to a number other than 200, the number of tasks in the shuffle stage will reflect that new value.

`Adaptive Query Execution (AQE)`: In Spark 3.0 and later, Adaptive Query Execution (AQE) is an optimization that can dynamically adjust the number of shuffle partitions during runtime. AQE can coalesce small shuffle partitions into larger ones, effectively reducing the number of tasks if the data distribution allows for it. If AQE is enabled and determines that fewer partitions are optimal, you will see fewer than 200 tasks.

`Data Volume and Distribution`: If the amount of data being processed is very small, or if the data is highly skewed (meaning a few keys have a disproportionately large amount of data), Spark might not utilize all 200 partitions efficiently, or AQE might optimize the partition count.

`Coalesce or Repartition before GroupBy`: If a coalesce or repartition operation was performed on the DataFrame immediately before the groupBy, it could explicitly set the number of partitions, overriding the default shuffle partition setting for that specific operation.

`Specific Optimization Strategies`: Certain optimization strategies or custom partitioning schemes might be in place that influence the number of partitions used during the groupBy operation, leading to a task count different from 200.
